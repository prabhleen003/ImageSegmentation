```python
import matplotlib as mpl
import matplotlib.pyplot  as plt
from sklearn.cluster import KMeans
```

```python
image = mpl.image.imread(r"C:\Users\prabh\OneDrive\Desktop\APPLE.jpeg")
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
#flatten the data
```

```python
X= image.reshape(-1,3)
```

```python
X.shape
```

```python
wscc =[]
for i in range(1,11):
    kmeans = KMeans(init='k-means++',n_clusters=i,)
    kmeans.fit(X)
    wscc.append(kmeans.inertia_)
```

```python
import seaborn as sns
sns.set()

plt.plot(range(1,11),wscc)
plt.title("the elbow point graph")
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
#n=3 is the cluster number
```

```python
kmeans = KMeans(n_clusters=3,n_init=10)
Y = kmeans.fit(X)
```

```python
segmented = kmeans.cluster_centers_[kmeans.labels_]
```

```python
segmented= segmented.reshape(image.shape)
```

```python
plt.imshow(segmented/255)
```
